### V - MATRICE DE CORRELATION - MC

````python
import pandas as pd
from scipy.stats import skew, kurtosis

data = df3.copy()

# Calculer la matrice de corrélation
corr = data.corr()

# Définir un seuil de corrélation
threshold = 0.8

# Initialiser une liste vide pour stocker les variables corrélées
correlated_vars = []

# Parcourir chaque colonne de la matrice de corrélation
for i in range(len(corr.columns)):
    for j in range(i):
        
        if (abs(corr.iloc[i, j]) > threshold) and (abs(corr.iloc[i, j]) != 1):
        # if abs(corr.iloc[i, j]) > threshold :
            
            correlated_vars.append((corr.columns[i], corr.columns[j]))

# Initialiser une liste vide pour stocker les variables sélectionnées
selected_vars = []
cols_to_delete = []

# Pour chaque paire de variables corrélées
for var1, var2 in correlated_vars:
    # Calculer la skewness et la kurtosis pour chaque variable    
    skewness1 = skew(data[var1])
    kurtosis1 = kurtosis(data[var1])
    
    skewness2 = skew(data[var2])
    kurtosis2 = kurtosis(data[var2])

    # Sélectionner la variable avec la meilleure distribution
    if (skewness1 < skewness2) and (kurtosis1 < kurtosis2):
        selected_vars.append(var1)
        cols_to_delete.append(var2)
        
    else:
        selected_vars.append(var2)
        cols_to_delete.append(var1)


# Visualisations --------------------------------------------------------------------
print("Nombre total de couple de variable corrélées (Mat. Corr): " + str(len(correlated_vars)))
print("\n")
for i in range(len(correlated_vars)):
    print(correlated_vars[i])


print("\n")
print("Variable corrélées avec une meilleur distribution (Mat. Corr): " + str(len(selected_vars)))
print("\n")
print(selected_vars)

#------------------------------------------------------------------------------------
# Isolation des variables corrélées à supprimer
print("\n")
print("Liste des variables à faible corrélation 'df3' après (Mat. Corr): " + str(len(selected_vars)))
for i in range(len(cols_to_delete)):
    print(cols_to_delete[i])
````

#### (MC) - Sélection des variables à supprimer - Résultats de la Matrice de Corrélation 

````python
import pandas as pd

### Ensembles avec doublons           
brut_var_A = pd.DataFrame(selected_vars)
brut_var_B = pd.DataFrame(cols_to_delete)

### Sans doublons(set : enleve les doublons) 
var_corr_A_res = pd.DataFrame(list(set(selected_vars)))
var_corr_B_res = pd.DataFrame(list(set(cols_to_delete))) 



## Listes des variables à amputer ----------
List_var_high_corr = list(set(selected_vars))
List_var_low_corr = list(set(cols_to_delete))

for x in List_var_high_corr:
    for y in List_var_low_corr:
        if x==y:
            List_var_low_corr.remove(x)
        
print("Liste des variables à faibles corrélations qu'on devra supprimer : \n",List_var_low_corr)
cols_to_drop = List_var_low_corr.copy()
tab_cols_to_drop = pd.DataFrame(cols_to_drop)
print("\n")


# Impression des résultats --------------------------------------------------------------------#
with pd.ExcelWriter("C:/Users/DASY/Downloads/" + "Matrice_Corrélation_Results.xlsx") as writer:
    var_corr_A_res.to_excel(writer, sheet_name="haute_corr", header=True)
    var_corr_B_res.to_excel(writer, sheet_name="faible_corr", header=True)
    
    tab_cols_to_drop.to_excel(writer, sheet_name="vars_to_delete", header=True)
    
    brut_var_A.to_excel(writer, sheet_name="haute_corr_avec_doublons", header=True)
    brut_var_B.to_excel(writer, sheet_name="faible_corr_avec_doublons", header=True)

print("Téléchargement résultats de matrice corrélation terminé !")
````

#### ACP - Contributions des variables, Qualités de représentations

````python
import pandas as pd
from sklearn.decomposition import PCA
from scipy.stats import skew, kurtosis

# Initialiser l'objet PCA
data = df4.copy()
pca = PCA()

# Calculer les composantes principales
pca.fit_transform(data)

# Récupérer les corrélations entre les variables et les composantes principales
corr = pd.DataFrame(pca.components_,columns=data.columns)

# Définir un seuil de corrélation
threshold = 0.8

correlated_vars = []

# Parcourir chaque colonne de la matrice de corrélation
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > threshold:
            correlated_vars.append((corr.columns[i], corr.columns[j]))

# Initialiser une liste vide pour stocker les variables sélectionnées
selected_vars = []
var_to_delete = []  # liste de variable à supprimer

# Pour chaque paire de variables corrélées
for var1, var2 in correlated_vars:
    
    # Calculer la skewness et la kurtosis pour chaque variable
    skewness1 = skew(data[var1])
    kurtosis1 = kurtosis(data[var1])
    
    skewness2 = skew(data[var2])
    kurtosis2 = kurtosis(data[var2])

    # Sélectionner la variable avec la meilleure distribution (inerties)
    if (skewness1 < skewness2) and (kurtosis1 < kurtosis2) :
        selected_vars.append(var1)
        var_to_delete.append(var2)   # Var à supprimer après corrélation
        
    else:
        selected_vars.append(var2)
        var_to_delete.append(var1)   # Var à supprimer après corrélation

        
# Visualisation des résultats --------------------------------------#
print("Total variables corrélées : " + str(len(correlated_vars)))

# Visualisation des couples de variables corrélées 2 à 2
print("\n")
print("Couple de variables corrélées 2 à 2 : " + str(len(correlated_vars)))
for i in range(len(correlated_vars)):
    print(correlated_vars[i])

print("\n")
print("Total variables sélectionnées corrélées avec de meilleures distributions : " + str(len(selected_vars)))
print(selected_vars)

# Isolation des variables corrélées à supprimer
print("\n")
print("Liste des variables à amputer de la table 'cat_var2': " + str(len(selected_vars)))
for i in range(len(var_to_delete)):
    print(var_to_delete[i])

# Récupération des variables sélectionnées
print("\n")
print("Visualisation des variables à garder : " + str(len(selected_vars)))
df_selected_vars = data[selected_vars]
df_selected_vars.head()
````
